```python
# Set the sample size to 70 (systematic sampling)
sample_size = 70
# Calculate the population size from sales_df
pop_size = len(sales_df)
# Calculate the interval
interval = pop_size // sample_size
# Systematically sample 70 rows
sales_sys_samp = sales_df.iloc[::interval]
# Print the sample
print(sales_sys_samp)
```

```python
# Add an index column to sales_df 
sales_df_id = sales_df.reset_index()

# Plot cost vs. index for sales_df_id
sales_df_id.plot(x='index', y='cost', kind='scatter')
plt.show()
```

```python
# Shuffle the rows of sales_df (systematic sampling with shuffling)
sales_shuffled = sales_df.sample(frac=1)

# Reset the row indexes and create an index column
sales_shuffled = sales_shuffled.reset_index(drop=True).reset_index()

# Plot cost vs. index for sales_shuffled
sales_shuffled.plot(x="index", y="cost", kind="scatter")
plt.show()
```

The above three code blocks together demonstrate systematic sampling and how to safely check it for hidden bias. The first block calculates an interval (population_size // sample_size) and pulls every n-th row from sales_df using .iloc[::interval]. The second block adds an index column via reset_index() and scatter-plots cost against row position to visually check whether the original data has any hidden order (a trend or clustering would signal danger for systematic sampling, while random noise would signal it's safe). The third block shuffles all rows with sample(frac=1), resets the index to match the new order, and re-plots cost vs. index to confirm the shuffle destroyed any pattern. Therefore, the intended workflow is to plot the original order first, and if it shows structure, shuffle before running the systematic sample in block one.

This code is doing cluster sampling — a sampling technique that's different from simple random sampling or stratified sampling. Here's the big picture:
The idea: Instead of sampling individual rows directly from the whole dataset, you first randomly pick a few entire groups (clusters), then only sample from within those chosen groups — ignoring the rest entirely.

The below code builds three 25%-sized samples of sales_df using different methods for comparison: sales_srs is a simple random sample of the whole dataset, sales_strat groups by category and takes 25% from each group so proportions match the population, and sales_clust instead randomly picks just 2 categories, filters to only those, and samples a fixed number of rows from each — representing those 2 categories thoroughly but excluding every other category entirely.

```python
import random

# Perform simple random sampling to get 0.25 of the population
sales_srs = sales_df.sample(frac=0.25, random_state=2022)

# Perform stratified sampling to get 0.25 of each category group
sales_strat = sales_df.groupby('category')\
    .sample(frac=0.25, random_state=2022)

# Create a list of unique category values
category_unique = list(sales_df['category'].unique())

# Randomly sample 2 unique category values
category_samp = random.sample(category_unique, k=2)

# Filter for category_samp and clear unused categories from category
category_condition = sales_df['category'].isin(category_samp)
sales_clust_prep = sales_df[category_condition].copy()
sales_clust_prep['category'] = sales_clust_prep['category'].astype('category').cat.remove_unused_categories()

# Perform cluster sampling on the selected group, getting 0.25 of sales_df split across the 2 clusters
n_per_group = (len(sales_df) // 4) // 2

sales_clust = sales_clust_prep.groupby("category", observed=False)\
    .sample(n=n_per_group, replace=True, random_state=2022)
# Print the sample
print(sales_clust)
```

```python
# Replicate the sampling code 1000 times
sample_means_1000 = []
for i in range(1000):
    sample_means_1000.append(
  		np.random.choice(list(range(1, 9)), size=5, replace=True).mean()
    )

# Draw a histogram of sample_means_1000 with 20 bins
plt.hist(sample_means_1000, bins=20)
plt.show()
```

The exact sampling distribution (Part 1) is the true, complete distribution of the sample mean, computed by enumerating all 32,768 possible dice combinations with no randomness or estimation involved — it's ground truth. The approximate sampling distribution (Part 2) instead estimates that same distribution using only 1,000 random samples, which is necessary in most real-world cases where enumerating every possible sample is computationally impossible; it looks similar in shape (bell-curved, centered around 4.5) but carries some random noise since it's based on a limited sample rather than every possible outcome. Comparing the two shows that random sampling, even without full enumeration, produces a distribution that closely approximates the true sampling distribution.

```python
# Generate sampling distributions of the mean cost, for sample sizes 5, 50, and 500
sampling_distribution_5 = []
for i in range(1000):
    sampling_distribution_5.append(
        sales_df['cost'].sample(n=5).mean()
    )

sampling_distribution_50 = []
for i in range(1000):
    sampling_distribution_50.append(
        sales_df['cost'].sample(n=50).mean()
    )

sampling_distribution_500 = []
for i in range(1000):
    sampling_distribution_500.append(
        sales_df['cost'].sample(n=500).mean()
    )

# Calculate the mean of the mean costs for each sampling distribution
mean_of_means_5 = np.mean(sampling_distribution_5)
mean_of_means_50 = np.mean(sampling_distribution_50)
mean_of_means_500 = np.mean(sampling_distribution_500)

# Print the results
print('Mean of 5:', mean_of_means_5)
print('Mean of 50:', mean_of_means_50)
print('Mean of 500:', mean_of_means_500)
```

As sample size grows (5 → 50 → 500), the mean of each sampling distribution should converge closer and closer to the true population mean of cost (sales_df['cost'].mean()), illustrating that the sample mean is an unbiased estimator — on average, across many repeated samples, it correctly estimates the population parameter, regardless of sample size. What does change with sample size is the spread/variance of the sampling distribution (smaller samples produce more variable estimates), which you'd typically visualize separately with histograms.

```python
# Calculate the std. dev. of the mean costs for each sampling distribution
sd_of_means_5 = np.std(sampling_distribution_5, ddof=1)
sd_of_means_50 = np.std(sampling_distribution_50, ddof=1)
sd_of_means_500 = np.std(sampling_distribution_500, ddof=1)

# Print the results
print('Standard Deviation of 5:', sd_of_means_5)
print('Standard Deviation of 50:', sd_of_means_50)
print('Standard Deviation of 500:', sd_of_means_500)
```

The amount of variation in the sampling distribution is related to the amount of variation in the population and the sample size. This is another consequence of the Central Limit Theorem.

```python
mean_cost_2000_samp = []

# Generate a sampling distribution of 2000 replicates
for i in range(2000):
    # Sample 500 rows and calculate the mean cost
    sample_mean = sales_df.sample(n=500, replace=False)['cost'].mean()
    mean_cost_2000_samp.append(sample_mean)

# Print the sampling distribution results
#print(mean_cost_2000_samp)
plt.hist(mean_cost_2000_samp)
plt.show()
```

This code builds a true sampling distribution of the mean cost: it loops 2,000 times, each time drawing a fresh, independent random sample of 500 rows directly from the full population (sales_df, without replacement), computes that sample's mean cost, and stores all 2,000 resulting means in a list. The final histogram then shows how these 2,000 sample means are distributed — which look roughly bell-shaped and centered near the true population mean, illustrating the Central Limit Theorem in action using genuine repeated sampling from the actual population, rather than resampling from a single existing sample (as bootstrapping would do).

```python
mean_cost_2000_boot = []

# Generate a bootstrap distribution of 2000 replicates
for i in range(2000):
    mean_cost_2000_boot.append(
        # Resample 500 rows and calculate the mean cost
        np.mean(sales_sample.sample(frac=1, replace=True)['cost'])
    )

# Print the bootstrap distribution results
#print(mean_cost_2000_boot)
plt.hist(mean_cost_2000_boot)
plt.show()
```

This code builds a bootstrap distribution of the mean cost: it loops 2,000 times, each time resampling the same size as sales_sample with replacement (so some rows repeat and others are left out), computes the mean cost of that resample, and stores all 2,000 resulting means in a list. The final histogram shows how these bootstrap means are distributed — unlike the previous exercise's sampling distribution, which drew fresh samples from the full population, this repeatedly resamples from just one existing sample, simulating the variability you'd expect from the population using only the data already in hand, which is the practical workaround used when you don't have access to draw new samples from the true population.

```python
# Calculate the true population mean cost
pop_mean = np.mean(sales_df['cost'])

# Calculate the original sample mean cost
samp_mean = np.mean(sales_sample['cost'])

# Calculate the sampling dist'n estimate of mean cost
samp_distn_mean = np.mean(mean_cost_2000_samp)

# Calculate the bootstrap dist'n estimate of mean cost
boot_distn_mean = np.mean(mean_cost_2000_boot)

# Print the means
print([pop_mean, samp_mean, samp_distn_mean, boot_distn_mean])
```

The population mean (94,369) is closely matched by the sampling distribution estimate (94,415), showing that repeated true sampling converges on the real value. But the original sample mean (89,685) was off by chance, and the bootstrap estimate (89,680) mirrors that same error rather than correcting it — showing bootstrapping can only reflect the sample it started from, not fix its bias.

```python
# Calculate the population std dev cost
pop_sd = sales_df['cost'].std(ddof=0)

# Calculate the original sample std dev cost
samp_sd = sales_sample['cost'].std()

# Calculate the sampling distribution estimate of std dev cost
samp_distn_sd = np.std(mean_cost_2000_samp, ddof=1) * np.sqrt(500)

# Calculate the bootstrap distribution estimate of std dev cost
boot_distn_sd = np.std(mean_cost_2000_boot, ddof=1) * np.sqrt(500)

# Print the standard deviations
print([pop_sd, samp_sd, samp_distn_sd, boot_distn_sd])
```

This is an important property of the bootstrap distribution. When you don't have all the values from the population or the ability to sample multiple times, you can use bootstrapping to get a good estimate of the population standard deviation.

The population standard deviation (51,514) and original sample standard deviation (58,725) are in a similar range, as expected. But the sampling distribution estimate (36,032) is meaningfully lower and the bootstrap estimate (130,263) is over twice the true value.

```python
from scipy.stats import norm

# Generate a 95% confidence interval using the quantile method
lower_quant = np.quantile(mean_cost_2000_boot, 0.025)
upper_quant = np.quantile(mean_cost_2000_boot, 0.975)
# Print quantile method confidence interval
print((lower_quant, upper_quant))
```

```python
# Find the mean and std dev of the bootstrap distribution
point_estimate = np.mean(mean_cost_2000_boot)
standard_error = np.std(mean_cost_2000_boot,ddof=1)

# Find the lower limit of the confidence interval
lower_se = norm.ppf(0.025, loc = point_estimate, scale = standard_error)

# Find the upper limit of the confidence interval
upper_se = norm.ppf(0.975, loc = point_estimate, scale = standard_error)

# Print standard error method confidence interval
print((lower_se, upper_se))
```
