```python
import pandas as pd
import numpy as np

# Sample 1000 rows from spotify_population
sales_df = pd.read_csv('Data/sales.csv')
sales_df.columns = sales_df.columns.str.strip()
sales_sample = sales_df.sample(n=100)

# Print the sample
print(sales_sample)
```

```python
print(sales_df.head())
```

```python
# Calculate the mean cost from sales_df (population)
mean_cost_pop = sales_df['cost'].mean()

# Calculate the mean cost from sales_sample
mean_cost_samp = sales_sample['cost'].mean()

# Print the means
print('Mean cost of population:', mean_cost_pop)
print('Mean cost of sample:', mean_cost_samp)
```

```python
# Create a pandas Series from the cost column of sales_df
cost_pop = sales_df['cost']
# Sample 100 values of cost_pop
cost_samp = cost_pop.sample(n=100)
# Calculate the mean of cost_pop
mean_cost_pop = np.mean(cost_pop)
# Calculate the mean of cost_samp
mean_cost_samp = np.mean(cost_samp)
print(mean_cost_pop)
print(mean_cost_samp)
```

```python
import matplotlib.pyplot as plt

# Visualize the distribution of cost with a histogram (population)
sales_df['cost'].hist(bins=np.arange(10000, 310001, 10000))  # bins of width 10,000
plt.show()

# Visualize the distribution of cost with a histogram (sample)
sales_sample['cost'].hist(bins=np.arange(10000, 310001, 10000))  # bins of width 10,000
plt.show()
```

```python
# Generate random numbers from a Normal(5, 2)
normals = np.random.normal(loc=5, scale=2, size=5000)

# Plot a histogram of normal values, binwidth 0.5
bins = np.arange(-2, 13.5, 0.5)
plt.hist(normals, bins = bins )
plt.show()
```

```python
# Generate random numbers from a Uniform(-3, 3)
uniforms = np.random.uniform(low=-3, high=3, size=5000)

# Plot a histogram of uniform values, binwidth 0.25
bins = np.arange(-3, 3.25, 0.25)
plt.hist(uniforms, bins=bins)
plt.show()
plt.show()
```

```python
# Sample 70 rows using simple random sampling and set the seed
cost_samp = sales_df.sample(n = 70, random_state = 18900217)

# Print the sample
print(cost_samp)
```

```python
# Set the sample size to 70 (systematic sampling)
sample_size = 70
# Calculate the population size from sales_df
pop_size = len(sales_df)
# Calculate the interval
interval = pop_size // sample_size
# Systematically sample 70 rows
sales_sys_samp = sales_df.iloc[::interval]
# Print the sample
print(sales_sys_samp)
```

```python
# Add an index column to sales_df 
sales_df_id = sales_df.reset_index()

# Plot cost vs. index for sales_df_id
sales_df_id.plot(x='index', y='cost', kind='scatter')
plt.show()
```

```python
# Shuffle the rows of sales_df (systematic sampling with shuffling)
sales_shuffled = sales_df.sample(frac=1)

# Reset the row indexes and create an index column
sales_shuffled = sales_shuffled.reset_index(drop=True).reset_index()

# Plot cost vs. index for sales_shuffled
sales_shuffled.plot(x="index", y="cost", kind="scatter")
plt.show()
```

```python
country_counts_pop = sales_df['country'].value_counts(normalize=True)
print(country_counts_pop)

sales_strat = sales_df.groupby('country')\
    .sample(frac=0.4, random_state=2022)

country_counts_strat = sales_strat['country'].value_counts(normalize=True)
print(country_counts_strat)
```

```python
# Proportion of sales by category
category_counts_pop = sales_df['category'].value_counts(normalize=True)
# Print category_counts_pop
print(category_counts_pop)

# Proportional stratified sampling for 40% of each category group
sales_strat = sales_df.groupby('category')\
    .sample(frac=0.4, random_state=2022)

# Calculate the category proportions from sales_strat
category_counts_strat = sales_strat['category'].value_counts(normalize=True)
# Print category_counts_strat
print(category_counts_strat)
```

```python
sales_eq = sales_df.groupby('country')\
    .sample(n=30, replace = True, random_state=2022)

category_counts_eq = sales_eq['country'].value_counts(normalize=True)
print(category_counts_eq)
```

```python
# Plot cost from sales_df as a histogram
sales_df['cost'].hist(bins=np.arange(0, 310001, 10000))
plt.show()

# Sample 400 sales records weighted by cost
sales_weight = sales_df.sample(n=400, weights="cost")

# Plot cost from sales_weight as a histogram
sales_weight.hist(column='cost', bins=np.arange(0, 310001, 10000))
plt.show()
```
