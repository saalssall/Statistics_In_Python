```python
import pandas as pd
import numpy as np

# Sample 1000 rows from spotify_population
sales_df = pd.read_csv('Data/sales.csv')
sales_df.columns = sales_df.columns.str.strip()
sales_sample = sales_df.sample(n=100)

# Print the sample
print(sales_sample)
```

```python
print(sales_df.head())
```

```python
# Calculate the mean cost from sales_df (population)
mean_cost_pop = sales_df['cost'].mean()

# Calculate the mean cost from sales_sample
mean_cost_samp = sales_sample['cost'].mean()

# Print the means
print('Mean cost of population:', mean_cost_pop)
print('Mean cost of sample:', mean_cost_samp)
```

```python
# Create a pandas Series from the cost column of sales_df
cost_pop = sales_df['cost']
# Sample 100 values of cost_pop
cost_samp = cost_pop.sample(n=100)
# Calculate the mean of cost_pop
mean_cost_pop = np.mean(cost_pop)
# Calculate the mean of cost_samp
mean_cost_samp = np.mean(cost_samp)
print(mean_cost_pop)
print(mean_cost_samp)
```

```python
import matplotlib.pyplot as plt

# Visualize the distribution of cost with a histogram (population)
sales_df['cost'].hist(bins=np.arange(10000, 310001, 10000))  # bins of width 10,000
plt.show()

# Visualize the distribution of cost with a histogram (sample)
sales_sample['cost'].hist(bins=np.arange(10000, 310001, 10000))  # bins of width 10,000
plt.show()
```

```python
# Generate random numbers from a Normal(5, 2)
normals = np.random.normal(loc=5, scale=2, size=5000)

# Plot a histogram of normal values, binwidth 0.5
bins = np.arange(-2, 13.5, 0.5)
plt.hist(normals, bins = bins )
plt.show()
```

```python
# Generate random numbers from a Uniform(-3, 3)
uniforms = np.random.uniform(low=-3, high=3, size=5000)

# Plot a histogram of uniform values, binwidth 0.25
bins = np.arange(-3, 3.25, 0.25)
plt.hist(uniforms, bins=bins)
plt.show()
plt.show()
```
