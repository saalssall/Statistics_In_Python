```python
# Set the sample size to 70 (systematic sampling)
sample_size = 70
# Calculate the population size from sales_df
pop_size = len(sales_df)
# Calculate the interval
interval = pop_size // sample_size
# Systematically sample 70 rows
sales_sys_samp = sales_df.iloc[::interval]
# Print the sample
print(sales_sys_samp)
```

```python
# Add an index column to sales_df 
sales_df_id = sales_df.reset_index()

# Plot cost vs. index for sales_df_id
sales_df_id.plot(x='index', y='cost', kind='scatter')
plt.show()
```

```python
# Shuffle the rows of sales_df (systematic sampling with shuffling)
sales_shuffled = sales_df.sample(frac=1)

# Reset the row indexes and create an index column
sales_shuffled = sales_shuffled.reset_index(drop=True).reset_index()

# Plot cost vs. index for sales_shuffled
sales_shuffled.plot(x="index", y="cost", kind="scatter")
plt.show()
```

The above three code blocks together demonstrate systematic sampling and how to safely check it for hidden bias. The first block calculates an interval (population_size // sample_size) and pulls every n-th row from sales_df using .iloc[::interval]. The second block adds an index column via reset_index() and scatter-plots cost against row position to visually check whether the original data has any hidden order (a trend or clustering would signal danger for systematic sampling, while random noise would signal it's safe). The third block shuffles all rows with sample(frac=1), resets the index to match the new order, and re-plots cost vs. index to confirm the shuffle destroyed any pattern. Therefore, the intended workflow is to plot the original order first, and if it shows structure, shuffle before running the systematic sample in block one.

```python
# Proportion of sales by country
country_counts_pop = sales_df['country'].value_counts(normalize=True)
# Print country_counts_pop
print(country_counts_pop)

# Proportional stratified sampling for 40% of each country group
sales_strat = sales_df.groupby('country')\
    .sample(frac=0.4, random_state=2022)

# Calculate the country proportions from sales_strat
country_counts_strat = sales_strat['country'].value_counts(normalize=True)
# Print country_counts_strat
print(country_counts_strat)
```

This code is doing proportional stratified sampling by country, and comparing the result back to the original population to confirm it worked as intended.

```python
# Proportion of sales by category
category_counts_pop = sales_df['category'].value_counts(normalize=True)
# Print category_counts_pop
print(category_counts_pop)

# Proportional stratified sampling for 40% of each category group
sales_strat = sales_df.groupby('category')\
    .sample(frac=0.4, random_state=2022)

# Calculate the category proportions from sales_strat
category_counts_strat = sales_strat['category'].value_counts(normalize=True)
# Print category_counts_strat
print(category_counts_strat)
```

This code demonstrates proportional stratified sampling — a technique that ensures your sample keeps the same relative proportions of each category as the original population.

```python
# Create a list of unique country values
country_pop = list(sales_df['country'].unique())
# Randomly sample four country values
country_samp = random.sample(country_pop, k=4)

# Filter for rows where country is in country_samp
country_condition = sales_df['country'].isin(country_samp)
sales_filtered = sales_df[country_condition]

# Remove categories with no rows
sales_filtered['country'] = sales_filtered['country'].astype('category').cat.remove_unused_categories()
# Randomly sample 10 sales records from each sampled country
sales_clust = sales_filtered.groupby("country").sample(n=10, random_state=2022)
# Print the sample
print(sales_clust)
```

This code is doing cluster sampling — a sampling technique that's different from simple random sampling or stratified sampling. Here's the big picture:
The idea: Instead of sampling individual rows directly from the whole dataset, you first randomly pick a few entire groups (clusters), then only sample from within those chosen groups — ignoring the rest entirely.

```python
import random

# Perform simple random sampling to get 0.25 of the population
sales_srs = sales_df.sample(frac=0.25, random_state=2022)

# Perform stratified sampling to get 0.25 of each category group
sales_strat = sales_df.groupby('category')\
    .sample(frac=0.25, random_state=2022)

# Create a list of unique category values
category_unique = list(sales_df['category'].unique())

# Randomly sample 2 unique category values
category_samp = random.sample(category_unique, k=2)

# Filter for category_samp and clear unused categories from category
category_condition = sales_df['category'].isin(category_samp)
sales_clust_prep = sales_df[category_condition].copy()
sales_clust_prep['category'] = sales_clust_prep['category'].astype('category').cat.remove_unused_categories()

# Perform cluster sampling on the selected group, getting 0.25 of sales_df split across the 2 clusters
n_per_group = (len(sales_df) // 4) // 2

sales_clust = sales_clust_prep.groupby("category", observed=False)\
    .sample(n=n_per_group, replace=True, random_state=2022)
# Print the sample
print(sales_clust)
```

```python
# Calculate the same thing for the stratified sample
mean_cost_strat = sales_strat.groupby('category', observed=False)['cost'].mean()
# Print the result
print(mean_cost_strat)
```
